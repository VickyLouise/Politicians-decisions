```python
import datetime
base = datetime.datetime.today()
numdays = 365 * 10
date_list = [(base - datetime.timedelta(days=x)).strftime("%Y-%m-%d") for x in range(0, numdays)]
```

```python
date_list_2013 = [date for date in date_list if '2013' in date]
date_list_2014 = [date for date in date_list if '2014' in date]
date_list_2015 = [date for date in date_list if '2015' in date]
date_list_2016 = [date for date in date_list if '2016' in date]
```
